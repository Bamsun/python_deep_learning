```python
power = pd.read_csv("power.csv")
power = power.set_index(pd.to_datetime(power["date"]))
power = power.drop("date",axis =1)
```

```python
def stockreturn(col1, col2):
    log1 = []
    log2 = []
    stockre = []
    for i in col1:
        log1.append(math.log(i))
    for i in col2:
        log2.append(math.log(i))
    for i in range(len(log1)):
        stockre.append((log1[i]-log2[i])*100)
    return stockre
```

```python
a = power["power demand"]
b = power["power demand"].shift(1)
power["return"]=stockreturn(a,b)
```

```python
def stockreturn(col):
    logstock = pd.DataFrame(np.log(col.values))
    logstock = logstock.diff()*100
    return logstock.values
```
